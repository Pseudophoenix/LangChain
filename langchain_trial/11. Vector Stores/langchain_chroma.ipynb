```python
from langchain_google_vertexai import ChatVertexAI,VertexAIEmbeddings
from langchain.vectorstores import Chroma
```

```python
from langchain.schema import Document

doc1=Document(
    page_content="Virat Kohli is one of the most successful and consistent batsmen in IPL history. Known for his aggressive batting style a",
    metadata={"team":"Royal Challengers Bangalore"}
)
doc2=Document(
    page_content="Rohit Sharma is the most successful captain in IPL history, leading Mumbai Indians to five titles. He's known for his cal ",
    metadata={"team":"Mumbai Indians"}
)
doc3=Document(
    page_content="MS Dhoni, famously known as Captain Cool, has led Chennai Super Kings to multiple IPL titles. His finishing skills, wicke",
    metadata={"team":"Chennai Super Kings"}
)
doc4=Document(
    page_content="Jasprit Bumrah is considered one of the best fast bowlers in T20 cricket. Playing for Mumbai Indians, he is known for his",
    metadata={"team":"Mumbai Indians"}
)
doc5=Document(
    page_content="Ravindra Jadeja is a dynamic all-rounder who contributes with both bat and ball. Representing Chennai Super Kings, his q",
    metadata={"team":"Chennai Super Kings"}
)
```

```python
docs=[doc1,doc2,doc3,doc4,doc5]
```

```python
vector_store=Chroma(
    embedding_function=VertexAIEmbeddings(model_name="gemini-embedding-001"),
    persist_directory="chroma_db",
    collection_name='sample'
)
```

```python
vector_store.add_documents(docs)
```

```python
vector_store.get(include=['embeddings','documents','metadatas'])
```

```python
vector_store.similarity_search(
    query="Who among these are bowlers?",
    k=1
)
```

```python
vector_store.similarity_search_with_score(
    query="Who among these are bowlers?",
    k=2
)
```

```python
# meta-data filtering
vector_store.similarity_search_with_score(
    query=" ",
    filter={"team":"Chennai Super Kings"}
)
```

```python
update_doc1=Document(
    page_content="Virat Kohli, the former captain of Royal Challengers Bangalore(RCB), is renowned for his aggressive leadership and consistency",
    metadata={"team":"Royal Challengers Bangalore"}
)
vector_store.update_document(document_id="b8de19f4-f60e-481d-b00d-e1c60852c4d7",document=update_doc1)
```

```python
# view documents
vector_store.get(include=['embeddings','documents','metadatas'])
```
