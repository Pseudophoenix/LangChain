## Step1a - Indexing (Document Ingestion)

```python
video_id="G42J2MSKyc8"
try:
    ytt_api=YouTubeTranscriptApi()
    transcript_list=ytt_api.fetch(video_id=video_id,languages=["en"])
    transcript=" ".join(chunk.text for chunk in transcript_list)
    # for transcript in transcript_list:
    print(transcript)
except TranscriptsDisabled:
    print("No captions available in this video")
```

## Step1b - Indexing (Text Splitting)

```python
splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
chunks=splitter.create_documents([transcript])
```

```python
chunks[0]
```

## Step1c & 1d - Indexing (Embedding Generation and Storing in Vector Store)

```python
# convert the chunks into embeddings and then  store those into FAISS vector store
embeddings=VertexAIEmbeddings(model_name="gemini-embedding-001")
vector_store=FAISS.from_documents(chunks,embeddings)
```

```python
vector_store.index_to_docstore_id
```

```python
vector_store.get_by_ids(['b191089a-14eb-4c9c-a36b-a2f8ecce10c6'])
```

## Step 2 - Retrieval

```python
# Retriever from the vector store 
retriever=vector_store.as_retriever(search_type="similarity",search_kwargs={"k":4})
```

```python
retriever
```

```python
# Retrieve from the vector store relevant vector entries that are most similar to the given query
retriever.invoke("What is crewai")
```

## Step 3 - Augmentation

```python
llm=ChatVertexAI(model_name="gemini-2.5-pro",temperature=1)
```

```python
prompt=PromptTemplate(
    template="""
You are a helpful assistant. Answer ONLY from the provided transcript context. If the context is insufficient, just say you don't know.
{context}
Question : {question}
""",
input_variables=['context','question']
)
```

```python
# Retrieve from the vector database the vectors that have the most similarity to the given query
question="Is the topic of AI discussed in this video? If yes then what was discussed"
retrieved_docs=retriever.invoke(question)
```

```python
retrieved_docs
```

```python
# Context text related to our query
context_text='\n\n'.join(doc.page_content for doc in retrieved_docs)
```

```python
# send the question and context_text fetched from the vector store to the prompt which will then be sent to the LLM
final_prompt=prompt.invoke({"context":context_text,"question":question})
```

## Step 4 - Generation

```python
# Semd the prompt having the question and context_text obtained from the vector store, to the LLM
answer= llm.invoke(final_prompt)
print(answer.content)
```

## Building a Chain

```python
from langchain_core.runnables import RunnableParallel,RunnableLambda, RunnablePassthrough
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
def format_docs(retrieved_docs):
    context_text="\n\n".join(doc.page_content for doc in retrieved_docs)
    return context_text
```

```python
parallel_chain=RunnableParallel(
    {
        'context':retriever|RunnableLambda(format_docs),
        'question':RunnablePassthrough()
    }
)
```

```python
# Output of the parallel chain will be context related to the query and the query itself
parallel_chain.invoke("Who is the author of the video")
```

```python
parser=StrOutputParser()
main_chain=parallel_chain|prompt|llm|parser
```

```python
main_chain.invoke("List the topics discussed in the video")
```
